```python
import os
import pandas as pd
import json
json_dir = '/.do/results/viper_gqa_sample2000_GQA_sample_2000/2024-01-06-all/json'
dir= os.listdir(json_dir)
result = pd.DataFrame()
for fname in dir:
    if fname.endswith('json'):
        name = fname.split('.')[0]
        sample_id = name
        with open(os.path.join(json_dir,fname)) as f:
            content = json.load(f)
            image_path = content['image_path']
            query = content['query']
            label = content['label']
            answer = content['answer']
            code_generate_time = content['code_generate_time']
            replan_remain_time = content['replan_remain_times']
            outputs=dict(
                sample_id=sample_id,
                query =query,
                image_path =image_path,
                label = label,
                answer =answer,
                code_generate_time = code_generate_time,
                replan_remain_time = replan_remain_time
            )
            result = result._append(outputs,ignore_index=True)
```

# CSV merge

```python
import pandas as pd
import os
import json
#get the label whole ours frame data
l_df = pd.read_excel('/.do/single_new_label.xlsx',sheet_name='single_new')
#get viper hard data with label
vl_df = pd.read_csv('/.do/viper_label.csv')
#get whole viper data
dir_path_2 = '/.do/results/viper_gqa_sample200_single_agent_viper_mode/2023-12-21-23:15/json'
dir2 = os.listdir(dir_path_2)
v_df = pd.DataFrame()
for fname in dir2:
    if fname.endswith('json'):
        name = fname.split('.')[0]
        sample_id = name
        with open(os.path.join(dir_path_2,fname)) as f:
            content = json.load(f)
            image_path = content['image_path']
            query = content['query']
            label = content['label']
            answer = content['answer']
            code_generate_time = content['code_generate_time']
            if code_generate_time>1:
                idx.append(sample_id)
            outputs=dict(
                sample_id=sample_id,
                query =query,
                image_path =image_path,
                label = label,
                answer =answer,
                code_generate_time = code_generate_time
            )
            v_df = v_df._append(outputs,ignore_index=True)
```

```python
vl_df = vl_df.rename(columns={'judge_result':'viper_judge_res'})
vl_df
```

# 得到代码生成次数大于2，3，4，5的各模式答案

```python
l_df_c2345 = l_df[l_df['code_generate_time']!=1]
merged_c2345_df = pd.merge(vl_df,l_df_c2345,on='sample_id',how='outer')
merged_c2345_df['viper_judge_res'].fillna(0,inplace=True)
merged_c2345_df['viper_judge_res'] = merged_c2345_df['viper_judge_res'].astype(int)
```

```python
l_df_c1 = l_df[l_df['code_generate_time']==1]
v_df['sample_id'] = v_df['sample_id'].astype(int)
merged_c1_df = pd.merge(v_df, l_df_c1, on='sample_id',how='right')
merged_c1_df['answer_x'] = merged_c1_df['answer_x'].astype(str)
merged_c1_df['answer_y'] = merged_c1_df['answer_y'].astype(str)
mask = merged_c1_df['answer_x'] == merged_c1_df['answer_y']
merged_c1_df.loc[mask,'viper_judge_res']=merged_c1_df.loc[mask,'judge_res'].astype(int)
merged_c1_df[mask]['viper_judge_res']
merged_c1_df.to_csv('merged_c1_df.csv')
```

# lavis评测工具使用

```python
import pandas as pd
from lavis.common.vqa_tools.vqa_eval import VQAEval
import json
data = pd.read_csv('/.do/final_df_without_acc.csv')
vqa_tool = VQAEval()
acc = []
v_acc = []
gt_file = '/.do/data/sample_data/code_vqa/cache/gqa/annotations/val_balanced_questions_sample2000.json'
fa2a={}
with open(gt_file,'r') as f:
    content = json.load(f)
    for item in content:
        fa = item['fullAnswer']
        a = item['answer']
        fa2a[fa] = a

for i in range(len(data)):
    d = data.iloc[i,:]
    gt_ans = d['label_y']
    gt_ans = fa2a[gt_ans]
    v_pred = d['answer_x']
    v_pred = str(v_pred)
    v_pred = vqa_tool.processPunctuation(v_pred)
    v_pred = vqa_tool.processDigitArticle(v_pred)

    pred = d['answer_y']
    pred = str(pred)
    pred = vqa_tool.processPunctuation(pred)
    pred = vqa_tool.processDigitArticle(pred)

    vqa_acc = 1 if pred == gt_ans else 0
    v_vqa_acc = 1 if v_pred == gt_ans else 0
    acc.append(vqa_acc)
    v_acc.append(v_vqa_acc)

accuracy = sum(acc) / len(acc) * 100
v_accuracy = sum(v_acc) / len(v_acc) * 100
```
