```python
import pandas as pd
data = pd.read_csv('./test/IBLIP.csv')
answer = data['pred_ans'].to_list()
wrong_list =[]
no_found_list = []
selected_sample_id1 = []
all_questions = data['question'].to_list()
accs = data['acc']
selected_questions = []

selected_questions_group_2 =[]
selected_questions_group_3 =[]
answer_group_2 = []
answer_group_3 = []
for sample_id,ans in enumerate(answer):
    ans = str(ans).lower()
    selected = False
    if 'sorry' in ans or 'unknown' in ans:
        selected = True
    elif "i don't know" in ans:
        selected = True
    elif "without the image" in ans:
        selected = True
    elif "without a specific image" in ans:
        selected = True
    elif "as an ai language model" in ans:
        selected = True
    elif 'no ' in ans:
        selected = True
    if selected:
        selected_sample_id1.append(sample_id)
        no_found_list.append(ans)
        selected_questions.append(all_questions[sample_id])
    else:
        if accs[sample_id] == 1 and len(selected_questions_group_2)<1000:
            answer_group_2.append(ans)
            selected_questions_group_2.append(all_questions[sample_id])
        if accs[sample_id] == 0 and len(selected_questions_group_3)<100:
            answer_group_3.append(ans)
            selected_questions_group_3.append(all_questions[sample_id])
```

```python
import json
from rich import print
from PIL import Image
from prompts.review_prompt import Get_Location_prompt_v3
from LLM.chatgpt_function_model import TemplateChatGpt,FindReviewFunction
# questioner = TemplateChatGpt(prompt_template=Get_Location_prompt_v3)
questioner = FindReviewFunction(prompt_template=Get_Location_prompt_v3)
with open('all_res.json','r') as f:
    all_res = json.load(f)
all_judge_res = {}
for id,content in all_res.items():
    res = content['res']
    query = content['query']
    objects = content['expected_result']
    ori_code = content['code']
    image_path = content['ori_image']
    code_list = ori_code.split('\n')
    objects = [o.replace('_patch','').replace('_patches','') for o in objects]
    res_group = res.split(';')
    for r,object in zip(res_group,objects):
        if 'None' in r or '[]' in r:
            image = Image.open(image_path)
            caption = InstructBlip(image,prompt=None)
            print('query:'+'\n' + query)
            print('caption:'+'\n' + caption)
            inputs = dict(
                query = query,
                object = object,
                caption = caption
            )
            res = questioner.parse(inputs,keep_history=False)
            all_judge_res[id] = res
```
